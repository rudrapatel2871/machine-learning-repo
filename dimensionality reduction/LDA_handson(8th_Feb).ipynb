```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/nba_final.csv')
```

```python
df['Pos2']=df['Pos2'].replace(np.nan,'NONE')
df['Pos2'].unique()
```

```python
df=df.fillna(0)
```

```python
df.isnull().sum()
```

```python
df=df.drop(['Player.x','Player_ID'],axis=1)
```

```python
df.columns
```

```python
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
```

```python
df['Pos1']=le.fit_transform(df['Pos1'])
df['Pos2']=le.fit_transform(df['Pos2'])
df['Tm']=le.fit_transform(df['Tm'])
df['Season']=le.fit_transform(df['Season'])
df['Conference']=le.fit_transform(df['Conference'])
df['Role']=le.fit_transform(df['Role'])
df['Play']=le.fit_transform(df['Play'])
```

```python
df
```

```python
x=df.drop(['Play'],axis=1)
y=df['Play']
```

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda=LinearDiscriminantAnalysis(n_components=1)
x=lda.fit_transform(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3, random_state=29)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
pred_dt = dt.predict(x_test)
```

```python
pred_dt
```

```python
y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import classification_report
cr_dt = classification_report(pred_dt, y_test)
print(cr_dt)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
```
