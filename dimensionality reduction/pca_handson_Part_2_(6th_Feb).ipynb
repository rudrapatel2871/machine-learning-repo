```python
from sklearn.decomposition import PCA
```

```python
# 0.95 is the % of info i would like to gain from the data after resducing the dimensions
# if 100%/1 is taken in the PCA() part you would observe overfitting
# so we retaining 95% of info even after dimensionality reduction or in other words compression
```

```python
pca=PCA(0.95)
x_pca=pca.fit_transform(x)
x_pca.shape
```
