```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
df['map']=encoder.fit_transform(df['map'])
df['bomb_planted']=encoder.fit_transform(df['bomb_planted'])
df['round_winner']=encoder.fit_transform(df['round_winner'])
```

```python
df
```

```python
x = df.drop(columns=['round_winner'])
y = df[['round_winner']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
ldaObj=LinearDiscriminantAnalysis()
```

```python
ldaObj.fit(x_train,y_train)
```

```python
ldaObj.transform(x_test)
```

```python
# We are going to calculate the LDA co-effecient, this will give us a score that will determine the importance associated.
```

```python
ldaCoef=np.exp(np.abs(ldaObj.coef_))
```

```python
ldaCoef
```

```python
ldaCoef=ldaCoef.flatten()
```

```python
ldaCoef
```

```python
# The number of features present in the dataset

num_feat=x.shape[1]
num_feat
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn=KNeighborsClassifier()
```

```python
kn.fit(x_train,y_train)
```

```python
y_pred=kn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print(accuracy_score(y_test,y_pred))
```
