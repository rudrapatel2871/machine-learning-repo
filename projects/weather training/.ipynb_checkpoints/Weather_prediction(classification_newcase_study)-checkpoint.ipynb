```python
cols=['MinTemp','MaxTemp','WindSpeed3pm','Rainfall','Evaporation','Sunshine','WindGustSpeed','WindSpeed9am','Humidity9am',
      'Humidity3pm','Pressure9am','Pressure3pm','Cloud9am','Cloud3pm','Temp9am','Temp3pm']

for i in range(len(cols)):
  df[cols[i]]=df[cols[i]].fillna(df[cols[i]].mean())
```

```python
df.isnull().sum()
```

```python
import random
```

```python
num_na=df['WindGustDir'].isna().sum()
num_na
```

```python
df.loc[df.WindGustDir.isna(),"WindGustDir"]=random.choices(['W','WNW','N','NNE','SW','ENE','SSE','NE','WSW','NNW','S','ESE','NW','E','SSW','SE'], k=num_na)
```

```python
wind_dir=df['WindDir9am'].isna().sum()
wind_dir
```

```python
df.loc[df.WindDir9am.isna(),"WindDir9am"]=random.choices(['W','WNW','N','NNE','SW','ENE','SSE','NE','WSW','NNW','S','ESE','NW','E','SSW','SE'], k=wind_dir)
```

```python
df.isnull().sum()
```

```python
wind_dir_3pm=df['WindDir3pm'].isna().sum()
wind_dir_3pm
```

```python
df.loc[df.WindDir3pm.isna(),"WindDir3pm"]=random.choices(['W','WNW','N','NNE','SW','ENE','SSE','NE','WSW','NNW','S','ESE','NW','E','SSW','SE'], k=wind_dir_3pm)
```

```python
df.isnull().sum()
```

```python
rain_today=df['RainToday'].isna().sum()
rain_today
```

```python
df.loc[df.RainToday.isna(),'RainToday']=random.choices(['Yes','No'],k=rain_today)
```

```python
df.isnull().sum()
```

```python
col=['MinTemp','MaxTemp','WindSpeed3pm','Rainfall','Evaporation','Sunshine','WindGustSpeed','WindSpeed9am','Humidity9am',
      'Humidity3pm','Pressure9am','Pressure3pm','Cloud9am','Cloud3pm','Temp9am','Temp3pm']
```

```python
for col_name in col:
  if(df[col_name].dtypes=='int64' or df[col_name].dtypes=='float64'):
    sns.boxplot(df[col_name])
    plt.xlabel(col_name)
    plt.ylabel('count')
    plt.show()
```

```python
Q1 = df['MinTemp'].quantile(0.25)
Q3 = df['MinTemp'].quantile(0.75)
IQR= Q3 - Q1
df = df[(df['MinTemp'] >= Q1 - 1.5 * IQR) & (df['MinTemp'] <= Q3 + 1.5 * IQR)]
```

```python
Q1 = df['WindGustSpeed'].quantile(0.25)
Q3 = df['WindGustSpeed'].quantile(0.75)
IQR= Q3 - Q1
df = df[(df['WindGustSpeed'] >= Q1 - 1.5 * IQR) & (df['WindGustSpeed'] <= Q3 + 1.5 * IQR)]
```

```python
for col_name in col:
  if(df[col_name].dtypes=='int64' or df[col_name].dtypes=='float64'):
    sns.boxplot(df[col_name])
    plt.xlabel(col_name)
    plt.ylabel('count')
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['RainToday']=le.fit_transform(df['RainToday'])
df['WindGustDir']=le.fit_transform(df['WindGustDir'])
df['WindDir9am']=le.fit_transform(df['WindDir9am'])
df['WindDir3pm']=le.fit_transform(df['WindDir3pm'])
```

```python
df=df.drop(columns=['row ID','Location',])
```

```python
df['RainTomorrow']=df['RainTomorrow'].astype(int)
```

```python
from sklearn.model_selection import train_test_split
x=df.drop('RainTomorrow',axis=1)
y=df['RainTomorrow']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=41)
```

```python
y_train.head()
```

```python
df
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
y_test_pred=lr.predict(x_test)
y_test_pred
```

```python
from sklearn import metrics
```

```python
print(metrics.accuracy_score(y_test,y_test_pred))
```
