The average monthly hours of a employee having 2 years experience is 167. Is it same for employees having more than 2 years experience?

Condition - Satisfactory level of a employee is from 0-0.5 and AvgMonthlyhours is from 130-200


````python

employee =df[(df.satisfactoryLevel>=0)&(df.satisfactoryLevel<=0.50)&(df.timeSpent_company>2)&(df.avgMonthlyHours>=130)&(df.avgMonthlyHours<=200)]
````

````python
employee.head()
````

````python
employee.shape
````

````python
# null hypothesis - avg monthly hours of a employee having 2 yrs exp and the employee more than that is same i.e, 167 hrs
# alternate hypothesis - avg monthly hours of a employee having 2 yrs exp doesnot have same avg monthly hours with more than 2 yrs exp employee
````

````python
# selecting a random sample
sample_size= 100
sample1= employee.sample(sample_size,random_state=0)
````

````python
sample1
````

````python
# population mean
pos_mean=167
````

````python
# sample mean
print(sample1["avgMonthlyHours"].mean())
````

````python
#The code uses the ttest_1samp function from the scipy.stats module to conduct a one-sample t-test on a sample of data.
from scipy.stats import ttest_1samp   #importing t test from stats library
````

````python

statistics,pvalue=ttest_1samp(sample1["avgMonthlyHours"],pos_mean)  #pos_mean is the population mean from which we have to compare
print(statistics,pvalue)


# sample1["avgMonthlyHours"] is a Pandas series that contains the sample data. This series is assumed to be normally distributed.
# pos_mean is the population mean that we are testing the sample mean against.

# statistics is the t-statistic for the test.
# pvalue is the two-tailed p-value for the test.
````

````python
# p value is less than 0.05
#reject null hypothesis,avg monthly hours is not 167 hours for employees having more than 2 yrs of exp
````

````python
#Now if we want to check upperlimit and lower limit value in confidence interval

# Degrees of freedom
n=len(employee["avgMonthlyHours"])
dof=n-1
print(n,dof)

# The degrees of freedom represent the number of values in the sample that are free to vary after estimating the sample mean.
# As the degrees of freedom increase, the t-distribution approaches the standard normal distribution (z-distribution).
# Therefore, larger sample sizes provide more reliable estimates of the population mean,
# resulting in larger degrees of freedom and a closer approximation to the normal distribution.
````

Students t-Test

The Students t-Test is a statistical hypothesis test for testing whether two samples are expected to have been drawn from the same population.

The t statistic calculated by the test can be interpreted by comparing it to critical values from the t-distribution. The critical value can be calculated using the degrees of freedom and a significance level with the percent point function (PPF).

What is PPF and CDF?


cdf() function calculates the probability for a given normal distribution value, while the . ppf() function calculates the normal distribution value for which a given probability is the required value. These are inverse of each other in this particular sense

The code uses the scipy.stats module in Python to compute the critical values (CVs) for a two-tailed t-test with a given significance level and degrees of freedom (dof).

````python
from scipy.stats import t #Import the t function from scipy.stats module.
````

````python
cv=t.ppf([alpha/2,1-alpha/2],dof)
print(cv)



# Call the ppf function of the t object to calculate the CVs.
# The ppf function stands for "percent point function", which is the inverse of the cumulative distribution function (CDF).
# The ppf function takes two arguments:


# An array of probabilities (in this case, [alpha/2,1-alpha/2]), which represent the area under the t-distribution curve
# that corresponds to the desired significance level.
# By dividing alpha by 2, we get the probability of each tail of the distribution,
# since the t-distribution is symmetrical around 0.
````

````python
# Null hypothesis - percentage of employee leaving the company having the exp of 2 years is same as employee having exp more than 2 yrs
# Alternate hypothesis - percentage of employee leaving the company having the exp of 2 years is different from the employee having exp more than 2 yrs
````

````python
from statsmodels.stats.proportion import proportions_ztest
````

````python
sample_size=100
````

````python
sample1=employee.sample(sample_size,random_state=0)
````

````python
p0=0.3
````

````python
statistic_oneprop,pvalue_oneprop=proportions_ztest(count=count,nobs=nobs,value=p0,alternative="two-sided",prop_var=False)
````

````python
print(statistic_oneprop,pvalue_oneprop)
````

Is the mean avg monthly hours of a employee having experience of 2-5 yrs is the same as that for employee having exp 6 - 10 yrs ?

````python
e1=df[(df.timeSpent_company>=2) & (df.timeSpent_company<=5)]
e2=df[(df.timeSpent_company>=6) & (df.timeSpent_company<=10)]
````

````python
e1.head()
````

````python
e1.shape
````

````python
e2.head()
````

````python
e2.shape
````

````python
s1=e1.sample(200,random_state=0)
s2=e2.sample(200,random_state=0)
````

````python
 import scipy
````

````python
scipy.stats.ttest_ind(s1["avgMonthlyHours"],s2["avgMonthlyHours"])
````

````python
# p value > 0.05 Hence accept null hypothesis
# The mean avg monthly hours of a employee having experience of 2-5 yrs is the same as that for employee having exp 6 - 10 yrs
````
