**Imputing the data for the missing records**

```python
for col_name in col_List:
  if df[col_name].dtypes == 'object':
    df[col_name] = df[col_name].fillna(df[col_name].mode()[0])
  else:
    df[col_name] = df[col_name].fillna(df[col_name].mean())
```

**Step 5: Visualizing the distribution**

* Visualizing using the histplot

```python
# distribution of data
for col_name in col_List:
  plt.hist(df[col_name])
  plt.xlabel(col_name)
  plt.ylabel('Count')
  plt.show()
```

* Visualizing using the distplot

```python
for col_name in col_List:
    if df[col_name].dtype in ['int64', 'float64']:
        sns.distplot(df[col_name])
        plt.xlabel(col_name)
        plt.ylabel('Count')
        plt.show()
```

* Visualizing using the kdeplot

```python
for col_name in col_List:
    if df[col_name].dtype in ['int64', 'float64']:
        sns.kdeplot(df[col_name])
        plt.xlabel(col_name)
        plt.ylabel('Count')
        plt.show()
```

**Step 6: Outliers Detection**

* Visualizing the outliers

```python
for col in col_List:
  if(df[col].dtypes == 'int64' or df[col].dtypes == 'float64' ):
    plt.boxplot(df[col])
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

* Treatment of Outliers

```python
df.describe()
```

```python
# BMI
Q1 = df.bmi.quantile(0.25)
Q3 = df.bmi.quantile(0.75)
Q1, Q3
```

```python
# IQR -> Interquartile Range
IQR = Q3 - Q1
```

```python
# upper limit and lower limit
# Lowerlimit
LL = Q1 - (1.5 * IQR)
UL = Q3 + (1.5 * IQR)
LL, UL
```

```python
plt.boxplot(df['bmi'])
plt.show()
```

```python
# treatment phase
df = df[(df.bmi >= LL) & (df.bmi <= UL)]
```

```python
plt.boxplot(df['bmi'])
plt.show()
```

```python
Q1 = df.past_consultations.quantile(0.25)
Q3 = df.past_consultations.quantile(0.75)
IQR = Q3 - Q1
LL = Q1 - (1.5 * IQR)
UL = Q3 + (1.5 * IQR)
df = df[(df.past_consultations >= LL) & (df.past_consultations <= UL)]
```

```python
plt.boxplot(df['past_consultations'])
plt.show()
```
